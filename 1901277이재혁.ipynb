```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([[13,19],[7,15],[11,13],[20,17],[16,9]])


data
```

```python
plt.scatter(data[:,0],data[:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,30)
plt.ylim(0,30)

rows, columns=data.shape
for i in range(rows):
     plt.annotate(i+1, (data[i,0],data[i,1]))
```

```python
scaler = StandardScaler()
scaler.fit(data)
data_std = scaler.transform(data)
print(data_std)
```

```python
plt.scatter(data_std[:,0], data_std[:,1])
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

rows, columns = data_std.shape
for i in range(rows):
  plt.annotate(i+1, (data_std[i,0], data_std[i,1]))  
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data_std)
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(data_std[:,0], data_std[:,1], c=kmeans.labels_, s=50)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

rows, columns = data_std.shape

for i in range(rows):
  plt.annotate(i+1, (data_std[i,0], data_std[i,1]))  
 
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1], 
            c='red', s=200, alpha=0.5);
for k in range(columns):
  plt.annotate(k, 
              (kmeans.cluster_centers_[k, 0], kmeans.cluster_centers_[k, 1]))
```

```python
TWSS = {}
for k in range(1, 6): 
  model = KMeans(n_clusters=k)
  model.fit(data_std)
  TWSS[k] = model.inertia_

plt.plot(list(TWSS.keys()), list(TWSS.values()), marker='o')
plt.xlabel("Number of Clusters(K)")
plt.ylabel("TWSS")
plt.show()
```
